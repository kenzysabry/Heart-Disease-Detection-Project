```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import  accuracy_score, classification_report, confusion_matrix
```

```python
df=pd.read_csv("clean_data.csv")
df.head()
```

```python
x=df.drop('target', axis=1)
y=df['target']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
model = DecisionTreeClassifier(random_state=42)
param_grid = {
    'max_depth': [3, 5, 7,10,20],
    'min_samples_split': [2, 5, 10,15,20],
}

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=5,
    scoring='accuracy'
)

grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
predictions = best_model.predict(x_test)
print("Best Params:", grid_search.best_params_)
print("Accuracy:", accuracy_score(y_test, predictions))

```

```python
train_pred = best_model.predict(x_train)
test_pred = best_model.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Test Accuracy:", accuracy_score(y_test, test_pred))
```

```python
confusion_matrix(y_test, test_pred)
print(classification_report(y_test, test_pred))
```

```python

# comparison

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Expert System
def expert_predict(row):
    age = row.get('age', 0)
    thalach = row.get('thalach', 0)
    oldpeak = row.get('oldpeak', 0)
    exang = row.get('exang_1', 0)

    # Rules
    if age > 55 and oldpeak > 2:
        return 1
    elif exang == 1 and thalach < 120:
        return 1
    elif oldpeak > 3:
        return 1
    else:
        return 0

expert_preds = x_test.apply(expert_predict, axis=1)


#  Validation Set Evaluation 

# Accuracy Comparison 
print("\nDecision Tree Performance")
print("Accuracy:", accuracy_score(y_test, test_pred))
print("Precision:", precision_score(y_test, test_pred))
print("Recall:", recall_score(y_test, test_pred))
print("F1 Score:", f1_score(y_test, test_pred))

print("\nExpert System Performance")
print("Accuracy:", accuracy_score(y_test, expert_preds))
print("Precision:", precision_score(y_test, expert_preds))
print("Recall:", recall_score(y_test, expert_preds))
print("F1 Score:", f1_score(y_test, expert_preds))
```
